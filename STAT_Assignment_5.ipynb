```python
import pandas as pd
d=pd.read_csv('/diabetes_data_upload.csv')
print(d)

```

```python
d.dtypes
```

```python
features=d.columns
print(features)
```

```python
target=d.columns[-1]
print(target)
```

```python
mapping={'Female':1,'Male':0,'No':3,'Yes':4,'Positive':5,'Negative':6}
d=d.replace(mapping)
print(d)
```

```python
df=d-d.min()/(d.max()-d.min())
print(df)
```

```python
y=d["class"]
print(y)
```

```python
x=d.drop("class",axis=1)
print(x)
```

```python
x = x.replace({'Gender': {'Male': 0, 'Female': 1},
                   'Polyuria': {'Yes': 1, 'No': 0},
                   'Polydipsia': {'Yes': 1, 'No': 0},
                   'sudden weight loss': {'Yes': 1, 'No': 0},
                   'weakness': {'Yes': 1, 'No': 0},
                   'Polyphagia': {'Yes': 1, 'No': 0},
                   'Genital thrush': {'Yes': 1, 'No': 0},
                   'visual blurring': {'Yes': 1, 'No': 0},
                   'Itching': {'Yes': 1, 'No': 0},
                   'Irritability': {'Yes': 1, 'No': 0},
                   'delayed healing': {'Yes': 1, 'No': 0},
                   'partial paresis': {'Yes': 1, 'No': 0},
                   'muscle stiffness': {'Yes': 1, 'No': 0},
                   'Alopecia': {'Yes': 1, 'No': 0},
                   'Obesity': {'Yes': 1, 'No': 0}})
print(x)
```

```python
y = y.replace({'Positive': 0, 'Negative': 1})
print(y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for column in x.columns:
  plt.plot(x[column], y)
  plt.xlabel(column)
  plt.ylabel('class')
  plt.title(f'{column} vs class')
  plt.show()
```

```python
x.isnull().sum()
x.fillna(0, inplace=True)
print(x)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = MinMaxScaler()
x_norm = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
print(x_norm)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=30)
print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
```
